```python
from scipy.stats import f_oneway

grouped_probs = data.groupby("probs")

Fs = []
Ps = []
columns = ["probs", *[str(i) for i in range(6)]]

for prob, prob_group in grouped_probs:
    print(prob)
    length = len(eval(prob))

    F = [prob]
    P = [prob]
    for i in range(0, length + 1):
        ssize_groups = [g[1][str(i)].to_numpy() for g in prob_group.groupby("n")]
        test_res = f_oneway(*ssize_groups)
        F.append(test_res.statistic)
        P.append(test_res.pvalue)
        print(f"{i}: {test_res}")
    print()
    while len(F) < len(columns):
        F.append(None)
        P.append(None)
    Fs.append(F)
    Ps.append(P)

df_Fs = pd.DataFrame(Fs, columns=columns)
df_Ps = pd.DataFrame(Ps, columns=columns)
df_Fs.to_csv("ANOVA_n_F.csv", index=False, float_format="%.4g")
df_Ps.to_csv("ANOVA_n_P.csv", index=False, float_format="%.4g")
```

```python
data["n_cat"] = data["probs"].map(lambda prob: len(eval(prob)))
grouped_n = data.groupby("n")

Fs2 = []
Ps2 = []
columns = ["n", "n_cat", *[str(i) for i in range(6)]]

for n, n_group in grouped_n:
    print(f"begin {n=}")
    n_cat_groups = n_group.groupby("n_cat")

    for n_cat, n_cat_group in n_cat_groups:
        print(f"{n_cat=}")
        F2 = [int(n)]
        P2 = [int(n)]
        F2.append(n_cat)
        P2.append(n_cat)
        for i in range(0, n_cat + 1):
            prob_groups = [g[1][str(i)].to_numpy() for g in n_cat_group.groupby("probs")]
            if len(prob_groups) < 2:
                print("skipping because of there's only one group")
                continue
            test_res = f_oneway(*prob_groups)
            F2.append(test_res.statistic)
            P2.append(test_res.pvalue)
            print(f"{i}: {test_res}")
        print()
        while len(F2) < len(columns):
            F2.append(None)
            P2.append(None)
        Fs2.append(F2)
        Ps2.append(P2)
    print(f"end {n=}")
    print()

df_Fs2 = pd.DataFrame(Fs2, columns=columns)
df_Ps2 = pd.DataFrame(Ps2, columns=columns)
df_Fs2.to_csv("ANOVA_probs_F.csv", index=False, float_format="%.4g")
df_Ps2.to_csv("ANOVA_probs_P.csv", index=False, float_format="%.4g")
```
